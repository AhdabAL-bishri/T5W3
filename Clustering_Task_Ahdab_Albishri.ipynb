##1_Impute missing data

```python
df['Urban_or_Rural_Area'].fillna(df['Urban_or_Rural_Area'].mean(), inplace=True)
df['Accident_Severity'].fillna(df['Accident_Severity'].mean(), inplace=True)
```

## 2_Remove the outliers

```python
#Remove the outliers
data_outliers = df.select_dtypes(include=['float', 'int']).columns

for i in data_outliers:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[~((df[i] < lower_bound) | (df[i] > upper_bound))]
```

## 2.1 Drop the not important




```python
df = df.drop(['Number_of_Casualties','Light_Conditions'], axis=1)
```

```python
data_not_important = ['Pedestrian_Crossing-Human_Control', 'Pedestrian_Crossing-Physical_Facilities',
                     'Did_Police_Officer_Attend_Scene_of_Accident',
                     'Local_Authority_(Highway)', 'Road_Type', 'Road_Surface_Conditions',
                     'Weather_Conditions']
df = df.drop(data_not_important, axis=1)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

##3_Scale the data

```python
Data_Scaling= ['Location_Easting_OSGR', 'Location_Northing_OSGR', 'Longitude', 'Latitude', 'Number_of_Vehicles',  'Speed_limit']

scaler = StandardScaler()
df[Data_Scaling] = scaler.fit_transform(df[Data_Scaling])
```

##4_Reduce dimentions using PCA


```python
Int_data= ['Location_Easting_OSGR', 'Location_Northing_OSGR', 'Longitude', 'Latitude', 'Number_of_Vehicles', 'Speed_limit']
Int_data2 = df[Int_data]

pca = PCA(n_components=2)
PCAresult = pca.fit_transform(Int_data2)
pca_df = pd.DataFrame(data=PCAresult, columns=['PC1', 'PC2'])
```

##5_ One-Hot Encoding

```python
df_encoded = pd.get_dummies(df, columns=['Day_of_Week','Urban_or_Rural_Area'])
```

## Clustering
Apply K-Means clustering. Determine the optimal number of clusters and justify your choice.
* Find the `n_clusters` parameter using the elbow method.
* Train the model.

```python
wcss = []
for i in range(1,10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(pca_df)
    wcss.append(kmeans.inertia_)
```

## Evaluation
Evaluate the clustering result using appropriate metrics.


```python
EvaluationK = silhouette_score(pca_df, kmeans.labels_)
print("Silhouette Score:", EvaluationK)
```

## Plot the data points with their predicted cluster center

```python
plt.figure(figsize=(12, 8))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=kmeans.labels_, cmap='viridis', s=60, alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='pink', marker='*', s=400, label='Cluster Centers')
plt.xlabel('Principal Component 1', fontsize=14)
plt.ylabel('Principal Component 2', fontsize=14)
plt.title('Clustering Results', fontsize=16)
plt.legend(loc='best', fontsize=12)
plt.grid(True, alpha=0.5)
plt.show()
```
